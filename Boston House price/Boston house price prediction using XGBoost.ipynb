```python
import pandas as pd
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
```

```python
dataset = sklearn.datasets.load_boston()
dataset
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df['price'] = dataset.target
df
```

```python
df.isnull().any()
```

```python
df.describe()
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
scalar = StandardScaler()
X_scale = scalar.fit_transform(X)
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.25, shuffle=True)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
prediction_train = model.predict(X_train)
```

```python
model.score(X_train, Y_train)
```

```python
from sklearn import metrics

score1 = metrics.r2_score(Y_train, prediction_train)

score2 = metrics.mean_absolute_error(Y_train, prediction_train)

print('R2 score is : ', score1)
print('Mean absolute error: ', score2)
```

```python
prediction_test = model.predict(X_test)
```

```python
model.score(X_test, Y_test)
```

```python
score1t = metrics.r2_score(Y_test, prediction_test)

score2t = metrics.mean_absolute_error(Y_test, prediction_test)

print('R2 score is : ', score1t)
print('Mean absolute error: ', score2t)
```

```python
import matplotlib.pyplot as plt

plt.scatter(Y_train, prediction_train)
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.show()
```
